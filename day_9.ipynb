```python
class Person:
    def __init__(self):
        self.name = input("Enter your name: ")
        self.age = int(input("Enter your age: "))
        self.gender = input("Enter your gender: ")

    def talk(self):
        print("Hi, I am", self.name)

    def vote(self):
        if self.age < 18:
            print("I am not eligible to vote")
        else:
            print("I am eligible to vote")


# Object creation (outside the class)
obj = Person()

# Method calls
obj.talk()
obj.vote()
```

```python
class Person:
    def __init__(self,n,a,g):
        self.name =n
        self.age =a
        self.gender =g

    def talk(self):
        print("Hi, I am", self.name)

    def vote(self):
        if self.age < 18:
            print("I am not eligible to vote")
        else:
            print("I am eligible to vote")


# Object creation (outside the class)
obj1 = Person("arjun",21,"male")
obj2 = Person("leela",17,"female ")

# Method calls
obj1.talk()
obj1.vote()

obj2.talk()
obj2.vote()
```
